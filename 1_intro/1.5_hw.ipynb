```python
import pandas as pd
import numpy as np
df = pd.read_csv('ds_salaries.csv', index_col= [0])
```

```python
#Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)
```

```python
df.groupby('job_title').count()
```

```python
#Какая средняя и медианная зарплата по группам вакансий?
```

```python
df_job_mean_salary = pd.DataFrame(round(df.groupby('job_title').salary.mean(),2))
df_job_mean_salary
```

```python
df_job_median_salary = pd.DataFrame(round(df.groupby('job_title').salary.median(),2))
df_job_median_salary
```

```python
#Какая средняя и медианная зарплата по каждому региону?
```

```python
df_company_mean_salary = pd.DataFrame( round(df.groupby('company_location').salary.mean(),2))
df_company_mean_salary
```

```python
df_company_median_salary = pd.DataFrame( round(df.groupby('company_location').salary.median(),2))
df_company_median_salary
```

```python
#Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?
```

```python
df_job_median_salary.sort_values(by='salary',ascending=False).head(1)
```

```python
df_job_mean_salary.sort_values(by='salary',ascending=False).head(1)
```

```python
#Какое процентное соотношение каждого региона по вакансиям от всех вакансий?
```

```python
x = df.groupby('company_location').count() * 100 /  len(df)
round(x['job_title'],2)
```

```python
#Какая корреляция уровня опыта от зарплаты?
```

```python
df_corr = df[['experience_level','salary']]
df_corr
```

```python
df_corr.loc[(df_corr.experience_level == 'EN'), 'experience_level'] = 1
df_corr.loc[(df_corr.experience_level == 'MI'), 'experience_level'] = 2
df_corr.loc[(df_corr.experience_level == 'SE'), 'experience_level'] = 3
df_corr.loc[(df_corr.experience_level == 'EX'), 'experience_level'] = 4
df_corr
```

```python
df_corr = df_corr.astype({'salary': np.int64})
df_corr = df_corr.astype({'experience_level': np.int64})
```

```python
print(df_corr.experience_level.corr(df_corr.salary))
```
